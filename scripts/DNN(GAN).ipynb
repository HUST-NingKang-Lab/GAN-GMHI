```python
# 配置参数
torch.manual_seed(1) #设置随机数种子，确保结果可重复
input_size = 542
hidden_size = 1083
num_classes = 2
num_epoched = 50
batch_size = 4347
learning_rate = 0.001
l=0
```

```python
class Net(nn.Module):
    def __init__(self,input_size,hidden_size,num_classes):
        super(Net,self).__init__()
        self.fc1=nn.Linear(input_size,hidden_size)
        self.relu=nn.ReLU()
        self.fc2=nn.Linear(hidden_size,num_classes)

    def forward(self, x):
        out=self.fc1(x)
        out=self.relu(out)
        out=self.fc2(out)
        return out
net=Net(input_size,hidden_size,num_classes)
```

```python
print(net)
```

```python
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(net.parameters(),lr=learning_rate)
```

```python
NH={'Healthy':[1], 'Nonhealthy':[0],'Unhealthy':[0]}
```

```python
val_dataset = [ val.iloc[i:i+1,1:].values for i in range(0,val.shape[0])]
```

```python
val1_dataset = [ val1.iloc[i:i+1,1:].values for i in range(0,val1.shape[0])]
```

```python
val_labels = val['Type']
```

```python
val1_labels = val1['Type']
```

```python
def test(val_dataset,val_labels):
    correct=0
    total=0
    for i,data in enumerate(val_dataset):#test set批处理
        datum = Variable(torch.Tensor(data))
        label=torch.LongTensor(NH[val_labels[i]])
        outputs=net(datum)
        #print(outputs)
        _,predicted=torch.max(outputs.data,1)#预测结果
        #print(label,predicted)
        total+=label.size(0)#正确结果
        correct+=(predicted==label).sum()#正确结果总数
    print('Accuracy of the network: %.2f %%'%(100*correct.item()/total))
```

```python
for epoch in range(num_epoched):
    for i, data in enumerate(dataset):#批处理
        #convert torch tensor to Variable
        datum = Variable(torch.Tensor(data))
        label=torch.LongTensor(NH[labels[i]])
        #forward+backward+optimize
        optimizer.zero_grad()#zero the gradient buffer梯度清零，以免影响其他batch
        outputs=net(datum)#前向传播
        loss=criterion(outputs,label)#loss
        loss.backward()#后向传播，计算梯度
        optimizer.step()#梯度更新

        if(i+1)%100==0:
            print('Epoch [%d/%d],Step[%d,%d],LossSum:%.6f'%(epoch+1,num_epoched,i+1,len(dataset)//batch_size,l))
            l=0
            #print(outputs,label)
        else:
            l=l+loss.item()
    print('Train set: ')
    test(dataset,labels)
    print('Test set: ')
    test(val_dataset,val_labels)
    print('Test set1: ')
    test(val1_dataset,val1_labels)
```
